```python
## Detect duplicate records based on name, age and city. State what each output reveals.
student_records.duplicated(subset=['name', 'age', 'city'])
```

```python
student_records['city'] = student_records['city'].str.lower()
```

```python
import numpy as np
student_records.loc[
  (student_records['age'] < 0) | (student_records['age'] > 120),
  'age'
] = None
mean_age = student_records['age'].mean()
student_records['age'] = student_records['age'].fillna(mean_age).astype(np.int64)
student_records
```

```python
student_records.drop_duplicates(
    subset=['name', 'age', 'city']
).reset_index()
```
