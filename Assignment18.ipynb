```python
## Importing required packages
import pandas as pd
```

```python
## Reading customer data
customer_data = pd.read_csv('/content/customer_data.csv')
print(customer_data.shape)
```

```python
## Exploring the data
customer_data
```

```python
customer_data.info()
```

```python
customer_data.describe()
```

```python
## Part 1 - Clean the data
customer_data['monthly_fee'] = pd.to_numeric(customer_data['monthly_fee'].str.replace('$', '').str.strip(), errors='coerce')
customer_data['signup_date'] = pd.to_datetime(customer_data['signup_date'], yearfirst=True, errors='coerce')
customer_data['cancellation_date'] = pd.to_datetime(customer_data['cancellation_date'], yearfirst=True, errors='coerce')
customer_data
```

```python
customer_data.info()
```

```python
## Part 2 - Indicate a churn indicator
customer_data['is_churned'] = customer_data['cancellation_date'].notna().astype('int')
customer_data
```

```python
## Part 3 - Calculate churn rate by subscription tier
churn_rate = customer_data.groupby('subscription_tier')['is_churned'].agg(
    churned_customers='sum',
    total_customers='count'
).assign(
    churn_rate_pct = lambda x: x['churned_customers'] / x['total_customers'] * 100
).reset_index().sort_values(
    'churn_rate_pct', ascending=False
)[['subscription_tier', 'churn_rate_pct']]
churn_rate
```

```python
## Part 4 - Identify a pattern
customer_data.groupby('is_churned')['total_watch_hours'].sum().reset_index()
```
