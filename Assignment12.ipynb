```python
## (a) Clean the orders DataFrame: remove duplicate rows (based on all columns) and reset the index. Print the shape before and after cleaning.
print(f"Before Removing Duplicates: {orders.shape}")
orders = orders.drop_duplicates(subset=orders.columns)
orders.reset_index()
print(f"After Removing Duplicates: {orders.shape}")
```

```python
## Clean the customers DataFrame: strip whitespace and apply title case to both the name and region columns. Print the cleaned DataFrame.
customers['name'] = customers['name'].str.strip().str.title()
customers['region'] = customers['region'].str.strip().str.title()
print(customers)
```

```python
## (c) Clean the products DataFrame: fill the missing category with 'Home', then convert the price column from strings like 'Rs. 500' to numeric floats (by removing the 'Rs. ' prefix). Print the cleaned DataFrame.
products['category'] = products['category'].fillna('Home')
products['price'] = pd.to_numeric(products['price'].str.replace('Rs.', '').str.strip()).astype(np.float32)
```

```python
products
```

```python
products.info()
```

```python
## (d) Merge all three DataFrames into a single DataFrame called report
# First merge orders (cleaned) with products (cleaned) on product_id
# Then merge the result with customers (cleaned) on customer_id
# Use left joins for both merges
# Add a new column total_value = quantity * price
# Print the shape and first 5 rows of report. (2.5 marks)

orders_with_products = pd.merge(orders, products, on='product_id', how='left')
report = pd.merge(orders_with_products, customers, on='customer_id', how='left')
report['total_value'] = report["quantity"] * report["price"]
print(report.shape)
print(report.head())
```

```python
report
```

```python
## (e) Produce a summary showing total revenue and number of orders per region, sorted by total revenue descending. Then create a pivot table with region as rows, category as columns, and total_value summed as the cell values (use fill_value=0). Print both outputs.

summary = report.groupby("region").agg(
    total_revenue = ("total_value", "sum"),
    total_orders = ("order_id", "count")
).reset_index()
print(summary)
category_region_sales = report.pivot_table(index='region', columns='category', values='total_value', aggfunc='sum', fill_value=0)
print(category_region_sales)
```
