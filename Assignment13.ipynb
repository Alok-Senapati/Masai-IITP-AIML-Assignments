```python
import pandas as pd
import numpy as np
```

```python
attendance = pd.DataFrame({
    'emp_id': ['E1', 'E2', 'E3', 'E4', 'E5', 'E6'],
    'department': ['HR', 'IT', None, 'Finance', 'IT', None],
    'days_present': [20, None, 22, 18, None, 19],
    'performance_score': [85, 90, None, None, 88, 92]
})
```

### Tasks:

(a) Identify the number of missing values in each column (2 marks)

(b) Drop rows where department is missing. Print the resulting DataFrame (2 marks)

(c) Fill missing values in days_present using the column mean (2 marks)

(d) Fill missing values in performance_score using forward fill (2 marks)

(e) After all cleaning steps, verify that no missing values remain (2 marks)

```python
attendance.head()
```

```python
print(f"Shape Before Dropping records with missing department: {attendance.shape}")
attendance.dropna(subset=['department'], inplace=True)
print(f"Shape After Dropping records with missing department: {attendance.shape}")
```

```python
attendance
```

```python
days_present_mean = attendance['days_present'].mean()
attendance.fillna({'days_present': days_present_mean}, inplace=True)
```

```python
attendance['performance_score'] = attendance['performance_score'].ffill()
```
